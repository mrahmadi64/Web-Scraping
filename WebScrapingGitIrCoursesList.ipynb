```python
import requests
from bs4 import BeautifulSoup

# آدرس URL سایت
url = "https://git.ir/courses/"

# هدرهای درخواست (شبیه‌سازی مرورگر)
headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36"
}

# درخواست GET به سایت با هدرها
response = requests.get(url, headers=headers)

# بررسی وضعیت پاسخ
if response.status_code == 200:
    # تجزیه HTML با استفاده از BeautifulSoup
    soup = BeautifulSoup(response.text, 'html.parser')
    
    # پیدا کردن تمام لینک‌های دوره‌ها
    course_links = soup.find_all('a', class_='text-secondary')
    
    # استخراج نام دوره‌ها
    courses = []
    for link in course_links:
        course_name = link.get('title')
        if course_name:
            courses.append(course_name)
    
    # نمایش نام دوره‌ها
    for i, course in enumerate(courses, 1):
        print(f"{i}. {course}")
else:
    print(f"خطا در دریافت اطلاعات: کد وضعیت {response.status_code}")
```

```python
import requests
from bs4 import BeautifulSoup

# تابع برای استخراج دوره‌ها از یک صفحه
def extract_courses_from_page(url, headers):
    response = requests.get(url, headers=headers)
    if response.status_code == 200:
        soup = BeautifulSoup(response.text, 'html.parser')
        course_links = soup.find_all('a', class_='text-secondary')
        courses = []
        for link in course_links:
            course_name = link.get('title')
            if course_name:
                courses.append(course_name)
        return courses
    else:
        print(f"خطا در دریافت اطلاعات از {url}: کد وضعیت {response.status_code}")
        return []

# هدرهای درخواست
headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36"
}

# آدرس پایه سایت
base_url = "https://git.ir/courses/"

# لیست تمام دوره‌ها
all_courses = []

# تعداد صفحات (در اینجا 3500 صفحه در نظر گرفته شده است)
total_pages = 3500

# حلقه برای پیمایش صفحات
for page in range(1, total_pages + 1):
    url = f"{base_url}?page={page}"
    print(f"در حال پردازش صفحه {page}...")
    courses = extract_courses_from_page(url, headers)
    if not courses:
        print(f"هیچ دوره‌ای در صفحه {page} یافت نشد. توقف عملیات.")
        break
    all_courses.extend(courses)

# نمایش تعداد کل دوره‌های جمع‌آوری شده
print(f"تعداد کل دوره‌های جمع‌آوری شده: {len(all_courses)}")

# ذخیره نتایج در یک فایل
with open("courses.txt", "w", encoding="utf-8") as file:
    for course in all_courses:
        file.write(course + "\n")

print("دوره‌ها با موفقیت در فایل courses.txt ذخیره شدند.")
```
